Slicing -> always from front to end...no matter index is negative or positive.

```python
nums = [10,20,30,40,50,60]

# Basic Slicing
nums[1:4]      # [20, 30, 40] → from index 1 to 3
nums[:3]       # [10, 20, 30] → from start to index 2
nums[3:]       # [40, 50, 60] → from index 3 to end

# Using Negative Indexes
nums[-3:]      # [40, 50, 60] → last 3 elements
nums[:-3]      # [10, 20, 30] → all except last 3
nums[-4:-1]    # [30, 40, 50] → from index -4 to -2

# Step Slicing (Every nth Element)
nums[::2]      # [10, 30, 50] → every 2nd element
nums[1::2]     # [20, 40, 60] → every 2nd element starting from index 1
nums[::-1]     # [60, 50, 40, 30, 20, 10] → reversed list

# Backward Slicing with Step
nums[4:1:-1]   # [50, 40, 30] → from index 4 to 2 (reverse)
nums[1::-1]    # [20, 10] → backward from index 1
nums[3:0:-2]   # [40, 20] → from index 3 to 1, step -2

# Entire List
nums[:]        # [10, 20, 30, 40, 50, 60] → full copy of list
nums[::]       # same as above (default step is 1)
```

1. Python treats values as True or False when used in conditions.
   * **Truthy:** Non-zero numbers, non-empty strings, lists, tuples, dictionaries, etc.
   * **Falsy:** 0, "", [], {}, (), None, False
2. **f-strings:** f-strings (formatted string literals) allow you to embed variables and expressions directly inside strings using {}.


```python
# --- Demo: Truthiness----
samples = [0, 1, "", "text", [], [1], {}, {"k": 1}, None]

for item in samples:
  if item:
    print(repr(item), "-> Truthy")
  else:
    print(repr(item), "-> Falsy")

# ---Demo: f-strings ---
name = "Sarafat"
score = 95
print(f"\n {name} scored {score}")
```

Problem: Analyze Order Data
Given the following nested dictionary representing a customer order, extract and print the following information:

The customer's full name.
The total price of all items in the order.
A list of the names of all items in the order.

```python
order_data = {
    "customer": {
        "first_name": "Bob",
        "last_name": "Johnson"
    },
    "items": [
        {"item_name": "Book",
         "price": 25.99
         },
        {"item_name": "Notebook",
         "price": 4.50
         },
        {"item_name": "Pen",
         "price": 1.20
         }
    ]
}


# Answer
print("Customer's Full Name: ", f"{order_data['customer']['first_name']} {order_data['customer']['last_name']}")  # using single qoute inside the f-string is good practice, as with double qoutes python gets confused
print("Total price: ", sum(item["price"] for item in order_data['items']))
print("Item List: ", [item["item_name"] for item in order_data['items']])
```

Comprehensions----

```python
# Demo: Comprehensions ----

temp = [35, 40, 28, 32, 30]

farhen = [item*(9/5) + 32 for item in temp]
print(f"farhen: {farhen}")

farhen1 = [item*(9/5) + 32 for item in temp if item % 2 == 0]
print(f"farhen_1: {farhen1}")

squares = [x**2 for x in range (6)]
print(f"squares: {squares}")

square_map = {x: x**2 for x in range(5)}
print(f"square_map: {square_map}")

unique_initials = {item[0].lower() for item in ["Alice", "Bob", "alex", "Beta"]}
print(f"unique_initials: {unique_initials}")
```
